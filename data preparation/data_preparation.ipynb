# Import Dataset

```python
data_url = 'https://raw.githubusercontent.com/Kenari-App-Bangkit-Capstone-Project/Kenari-MachineLearning/main/dataset/mental%20health%20conversation%20dataset%20(ind%20version).json'
urllib.request.urlretrieve(data_url, 'mental_health_conversation_dataset.json')
```

## Read Dataset

```python
with open('mental_health_conversation_dataset.json', 'r') as f:
    data = json.load(f)

data = pd.DataFrame(data['intents'])
data
```

# Dataset #1
Mentransformasikan setiap patterns dan responses

```python
dic = {"tag":[], "patterns":[], "responses":[]}
for i in range(len(data)):
    tag = data[data.index == i]['tag'].values[0]
    patterns = data[data.index == i]['patterns'].values[0]
    responses = data[data.index == i]['responses'].values[0]
    for j in range(len(patterns)):
        for i in range(len(responses)):
            dic['tag'].append(tag)
            dic['patterns'].append(patterns[j])
            dic['responses'].append(responses[i])

df = pd.DataFrame.from_dict(dic)
df
```

# Dataset #3
Mentransformasikan setiap patterns, dan melakukan preproceesing data menggunakan metode case folding, stemming dan stop words removal

```python
df3 = df2.copy()
df3
```

### Case folding

```python
import re

def case_folding(data):
  temp_text = []

  for text in data:

    text = re.sub(r"://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*,]|(?:%[0-9a-fA-F][0-9a-fA-F]))+", "", text)
    text = re.sub("[^a-zA-Z^']", " ", text)
    text = re.sub('\n', '', text)
    text = text.strip()
    text = text.lower();
    temp_text.append(text)
  return temp_text
```

```python
df3['patterns']=case_folding(df3['patterns'])
df3.head()
```

### Tokenize

```python
import nltk
nltk.download('punkt')

def tokenized(data):
  return data.apply(nltk.word_tokenize)
```

```python
df3['patterns']=tokenized(df3['patterns'])
df3.head()
```

### Stemming

```python
!pip install Sastrawi
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(data):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()

  return [stemmer.stem(patterns) for patterns in data]
```

```python
df3['patterns'] = df3['patterns'].apply(stemming)
df3.head()
```

### Stop words removal

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

def removeStopWords(data):
  stop_words = set(stopwords.words('indonesian'))
  return data.apply(lambda x: [item for item in x if item not in stop_words])
```

```python
df3['patterns']=removeStopWords(df3['patterns'])
df3.head()
```

### Finalisasi

```python
df3['patterns'] = df3['patterns'].apply(lambda x: ' '.join(x))
df3.head()
```

```python
df3.to_csv("3. conversation_mental_health_IND with preprocessing data (for classification).csv")
```
